## Phys2 related
Read in the data and draw bar graphs for corresponding choices.

```python
df_phys2 = pd.read_csv('survey/phys2.csv', header=None).T
df_phys2.columns = ['April 30', 'May 12', 'June 09']
df_phys2_econ = df_phys2.iloc[[5, 7, 10], :]
index_econ_related = ['C6', 'C8', 'C11']

df_phys2_fun = df_phys2.iloc[[1,14,15,16,17,18], :]
index_fun = ['C2', 'C15', 'C16', 'C17', 'C18', 'C19']
```

```python
pd.DataFrame(data=df_phys2_fun.values, columns=df_phys2_fun.columns, index=index_fun).plot.bar(rot=0)
```

```python
pd.DataFrame(data=df_phys2_econ.values, columns=df_phys2_econ.columns, index=index_econ_related).plot.bar(rot=0)
```
